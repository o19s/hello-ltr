```python
config = [
    {
        "store": "title",
        "name" : "title_phrase",
        "class" : "org.apache.solr.ltr.feature.SolrFeature",
        "params" : {
          "q" : "title:\"${keywords}\"^=1.0"
        }
    }
    
]

from ltr import setup_ltr
setup_ltr.run(config=config, featureset='title')
```

```python
from ltr import collectFeatures
trainingSet = collectFeatures.trainingSetFromJudgments(judgmentInFile='data/title_judgments.txt', 
                                                       trainingOutFile='data/title_judgments_train.txt', 
                                                       featureSet='title')
```

```python
from ltr import train
trainLog = train.run(trainingInFile='data/title_judgments_train.txt',
                     metric2t='DCG@10',
                     featureSet='title',
                     modelName='title')

print()
print("Impact of each feature on the model")
for ftrId, impact in trainLog.impacts.items():
    print("{} - {}".format(ftrId, impact))
```

```python
config = [
    {
        "store": "title_fuzzy",
        "name" : "title_phrase",
        "class" : "org.apache.solr.ltr.feature.SolrFeature",
        "params" : {
          "q" : "title:\"${keywords}\"^=1.0"
        }
    },
    {
        "store": "title_fuzzy",
        "name" : "title_fuzzy",
        "class" : "org.apache.solr.ltr.feature.SolrFeature",
        "params" : {
          "q" : "title:(${fuzzy_keywords})^=1.0"
        }
    }
]


from ltr import setup_ltr
setup_ltr.run(config=config, featureset='title_fuzzy')
```

```python
from ltr import collectFeatures
trainingSet = collectFeatures.trainingSetFromJudgments(judgmentInFile='data/title_judgments_fuzzy.txt', 
                                                       trainingOutFile='data/title_judgments_fuzzy_train.txt', 
                                                       featureSet='title_fuzzy')


from ltr import train
trainLog = train.run(trainingInFile='data/title_judgments_train.txt',
                     metric2t='DCG@10',
                     featureSet='title_fuzzy',
                     modelName='title_fuzzy')

print()
print("Impact of each feature on the model")
for ftrId, impact in trainLog.impacts.items():
    print("{} - {}".format(ftrId, impact))
```

```python
from ltr import search
search.run("War Games", modelName='title_fuzzy')
```
